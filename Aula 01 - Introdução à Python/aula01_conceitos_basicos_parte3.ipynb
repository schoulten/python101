# Estruturas condicionais

```python
# Importar biblioteca numpy
import numpy as np

# Gerar um número aleatório (normal dist.)
num = np.random.normal()
num
```

```python
# Verificar se número é maior do que 1
if num > 1:
  print("É maior do que 1")
elif num > -1:
  print("Entre -1 e 1")
else:
  print("Menor ou igual a -1")
```

# Estruturas de repetição

```python
# While
while(False):
  print("hi")
```

```python
contador = 1
while(contador < 5):
  print("hi")
  contador += 1
```

```python
# For
for i in [23, 45, 56]: # lista é a repetição
  print("hi" + " " + str(i))
```

```python
# For com tupla
for i in (1, 3, 6):
  print("hi" + " " + str(i))
```

```python
# For com string
for i in "FURG":
  print(i)
```

```python
# For com sequência definida (range)
for i in range(4):
  print(i)
```

```python
# For com sequência definida (range)
for i in range(4, 9):
  print(i)
```

```python
# For nested
for i in range(2):
  for j in range(3):
    print(str(i) + str(j))
```
