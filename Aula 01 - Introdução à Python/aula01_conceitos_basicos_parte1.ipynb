### 3. Tipo sequências

#### 3.1 Strings

```python
# Strings
texto = "python PUC"
type(texto)
```

```python
# Acessar elementos dentro de strings
texto[0] # primeiro elemento
texto[5] # elemento 5
texto[-2] # acessar elementos do final para o início da string
```

```python
# Modificar strings
texto = "novo valor "
texto
```

#### 3.2 Listas


```python
# Listas
lista1 = [] # criar lista
type(lista1)
```

```python
# Inserir elementos na lista
lista1 = ["puc", "python", 1, 3.5, False, [1, 2, 3, 4]]
type(lista1)
```

```python
lista1
```

```python
# Adicionar elementos na lista
lista1.append("Fernando")
lista1
```

```python
# Adicionar elemento em posição específica
lista1.insert(3, "Inserção")
lista1
```

```python
# Acessar elemento de uma lista
lista1[3]
lista1[0]
lista1[-1]
lista1[-2][0]
```

```python
# Remover elementos da lista
lista1.remove("python") # Não é por posição do elemento na lista
lista1
```

```python
# Remover elementos da lista por posição
lista1.pop(3)
lista1
```

### 4. Tipo Set


```python
# Criar tipo set
set1 = set()
type(set1)
```

```python
# Adicionar elementos
set1 = set(["puc", '1', 1, "puc", 2.2, False, True]) #internamento o True é o número 1 e False é 0
set1
```

```python
# Inserir elementos
set1.add(8)
set1
```

```python
set1.add(1)
set1
```

```python
# Adicionar conjunto de valores ao set
set1.update([3, 1, "teste", "manoela", 32342])
set1
```

```python
# Acessar elementos em um set (transformar em uma lista pra poder acessar)
print(list(set1))
list(set1)[0]
```

```python
# Remover elementos do set
set1.remove("manoela")
set1
```
